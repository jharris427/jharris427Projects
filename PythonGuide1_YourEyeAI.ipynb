Now that we have opencv, we can `import` the libraries we need for our code:

```python
# import the below libraries. "as" indicates how will we will call the library
import numpy as np
from matplotlib import pyplot as plt

# import computer vision library
import cv2 as cv
```

As an example for this exercise, we will be using an image of a Husky. However, you can use an image of your choosing. For simplicity, I recommend that you download the image to the same folder in which you are saving your code.

```python
## image without file path
image_ = "Northeastern_Huskies.jpg"

# Define image path 
# This is the image's file path on my computer. If you downloaded the Husky image in the same file as your folder,
# there is no need to define the folders. All you need then is the file name, as shown above^
image_path = "/Users/jonathanharris/Documents/Python/Northeastern_Huskies.jpg"
```

Once our libraries are loaded, we can use pre-defined functions from the OpenCV library to <a href="https://www.geeksforgeeks.org/python-opencv-cv2-imread-method/">read</a> and <a href="https://docs.opencv.org/master/d4/d13/tutorial_py_filtering.html">blur</a> the image, as seen in the linked guides from online.

<i>Note:</i> I have taken most of the code directly from the guides so feel free to follow along there, too. My advice is to look up what each function() does as you go along. This way, you can understand what arguments go into each function. As you get more comfortable with a function, you can adjust and add more parameters/arguments that you want to explore.

```python
# load the image into a variable as a matrix
img = cv.imread(image_path)
```

```python
# use the matplotlib functions below to show the image before processing.
plt.subplot(121), plt.imshow(img), plt.title('Original Husky')

# the .shape method outputs the shape of the image array
print(img.shape)
```

Next, we would like to try <i>filtering</i> the image with convolution by using a normalized box filter. To complete this operation, we will first define a `numpy` array for our kernal. 

Within the `np.ones` function, we can decide on a shape for an array of ones and then divide by the pixel theoretical dimensions of the array.

Then, we can use the `cv.filter2D` function to filter the image using the defined kernel.

The remaining code uses `matplotlib`, as indicated by `plt`, to display our original image and results.

```python
# define kernel
kernel = np.ones((10,10),np.float32)/100

# call filter function
dst = cv.filter2D(img,-1,kernel)

# display images
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

We can also blur the image with a Gaussian blur function, built in to the cv2 library.

```python
# Call blur function
blur = cv.GaussianBlur(img,(5,5),0)

# display images
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```
